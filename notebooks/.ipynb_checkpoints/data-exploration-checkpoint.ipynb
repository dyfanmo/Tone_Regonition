## Introduction


Mandarin Chinese is a tonal language; each tone said can change the meaning of a word. There are five distinctive tones in Mandarin Chinese most native speaker can differentiate. The purpose of this project is to see if I can train a model to replicate this skill, separating each tone into different classes.

Due to the advancements in computer vision,  most audio classification projects visualise the data to improve performance. The best format to visualise audio data into for this project is a Mel Spectrogram. A Mel spectrogram is a combination of a spectrogram, visualising sound vibrations, and a Mel scale that transforms frequency content to represent what humans consider important. Humans can detect the difference between 50hz and 150hz sound but cannot between 15000hz and 14500hz. The Mel scale transforms linear frequencies to focus more on the differences between lower frequencies. Using this data, the model can learn to detect differences in a sound similar to the human ear, necessary for a project where I aim to replicate a human ability.

To build the data, I downloaded a text file, including the Chinese dictionary, used to extract each character into a list. https://api.forvo.com/ is the website used to request a human audio pronunciation of each word. Using a variety of words and different human pronunciations forces the model to learn patterns in the tones, not to differentiate the words of the speakers themselves. 

To clean the data, I sampled the audio duration standard deviations*2 away from the mean, removing all outliers. Google Speach is used Recognition to asses the quality of the audio, removing audio where more than one syllable is detected and predict the tone of the audio. If the speech recognition predicted tone matches the actual tone, the audio quality is suitable for training. Google Speach Recognition is not perfect, so I used the pigeon library to check each audio file manually. This process improves the quality of the data so it can learn the right patterns for classification. There are 1000 audio samples left, enough to train an accurate model.

Building the training data requires converting mp3 files to wave files, then using the Librosa library to convert wave data into Mel spectrograms. Then I can store each image pixel into a NumPy array that I can then convert into a PyTorch tensor.

There are three models used in this project:
1. SimpleCNN Model — composed of 3 convolutional layers
2. ComplexCNN  Model — including  7 convolutional layers, a dropout layer and uses batch normalisation
3. ResNet Model — using pretrained layers and is the deepest model out of the three

The ResNet model performed the best, achieving 84% accuracy score. Several factors can explain why this model outperformed the others:
* Using pre-trained layers can still achieve high performance on a small dataset. 
* The use of residual blocks allows me to train a deep model without diverging. 
* Using a deep model allows for a higher level of feature extraction. 

We can see the more complex the model, the better the results; the complexCNN outperformed the simpleCNN by nearly 30%. The models need to learn through a high level of feature to understand the patterns in each class.

There are two functions, one responsible for making the Mel spectrogram and the other scaling the spectrogram into an image.

A Mel spectrogram is a spectrogram with the Mel Scale as the y-axis. It can visualise the vibrations of sound and place more emphasis on the differences between lower frequencies.

```python
def get_melspectrogram_db(file_path, sr=None, n_fft=2048, hop_length=512, n_mels=128, fmin=20, fmax=8300, top_db=80):
    wav,sr = librosa.load(file_path,sr=sr)
    if wav.shape[0]<5*sr:
        wav=np.pad(wav,int(np.ceil((5*sr-wav.shape[0])/2)),mode='reflect')
    else:
        wav=wav[:5*sr]
    spec=librosa.feature.melspectrogram(wav, sr=sr, n_fft=n_fft,
              hop_length=hop_length,n_mels=n_mels,fmin=fmin,fmax=fmax)
    spec_db=librosa.power_to_db(spec,top_db=top_db)
    return spec_db


def spec_to_image(spec, eps=1e-6):
    mean = spec.mean()
    std = spec.std()
    spec_norm = (spec - mean) / (std + eps)
    spec_min, spec_max = spec_norm.min(), spec_norm.max()
    spec_scaled = 255 * (spec_norm - spec_min) / (spec_max - spec_min)
    spec_scaled = spec_scaled.astype(np.uint8)
    return spec_scaled
```

As I am using many epochs iteration, I am using the learning rate decay technique prevent the model from diverging as it gets closer to the minimum. 

```python
def setlr(optimizer, lr):
    for param_group in optimizer.param_groups:
        param_group['lr'] = lr
    return optimizer

def lr_decay(optimizer, epoch):
    if epoch%20==0:
        new_lr = learning_rate / (10**(epoch//20))
        optimizer = setlr(optimizer, new_lr)
        print(f'Changed learning rate to {new_lr}')
    return optimizer
```

## ResNet model


Our model is low in variance due to the size of the training data. The best way to combat this is to use a model with pre-trained and modify the output layer. I am using the ResNet that allows me to train a deep model without diverging. The ResNet is the most complex model out of the other two.

```python
resnet_model = resnet34(pretrained=True)
resnet_model.fc = nn.Linear(512,4)
resnet_model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
```

```python
learning_rate = 2e-5
optimizer = optim.Adam(resnet_model.parameters(), lr=learning_rate)
epochs = 50
loss_fn = nn.CrossEntropyLoss()
train_losses = []
valid_losses = []
```

```python
train_model(resnet_model, loss_fn, train_loader, valid_loader, epochs, optimizer, train_losses, valid_losses, lr_decay)
```

The ResNet model performed the best achieving the lowest validation loss and the best validation accuracy score of 85%. The deeper the model, the better the performance without the risk of overfitting the training data.
